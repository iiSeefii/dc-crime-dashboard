```python
import pandas as pd
import numpy as np
import ftfy
```

```python
data = pd.read_excel("euro_lineups euro_coaches euro_summary.xlsx")
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
```

```python
pd.Series(list(data["name"].str.replace(r"[^A-Za-z0-9]","",regex=True).sum())).value_counts()

pd.Series([
    c for name in data["name"].dropna().astype(str)
    for c in name
    if not c.isalnum()
    ]).value_counts()
```

```python
data["name"] = data["name"].apply(ftfy.fix_text)
```

```python
df = pd.DataFrame({
    'Symbols:': ['«','„','¦','–',"'",'°','¯','†','®','¨','¢','´','‰',"أ©","›","أ¨","أª","أ«","أ¡","أ³","أ­","أº","أ±","أ¼","أ¶","أ¤","أŸ","أ¸","أ¥","أ§","أŒ","œ","إ½","إ¾","إ،","ؤچ","ؤ‡","إ‘", "‘", "‚","إ±","إ™","أ°","أ¾","–","‘","’"],
    'Letters:': ['ë','ń','æ','Ö',"$",'İ','ů','Ć','î','è','Ţ','ô','É',"é","é", "è", "ê", "ë", "á", "ó", "í", "ú", "ñ", "ü", "ö", "ä", "ß", "ø", "å", "ç", "Œ", "œ", "Ž", "ž", "š", "č", "ć", "ő", "ő", "ő", "ű", "ř", "ğ", "ş", "-", "'", "'"]
})

mapping = dict(zip(df['Symbols:'], df['Letters:']))

mapping.update({
    "إ ": "Š",
    "إ ": "Š",
})

df
```

```python
cols = ['name', 'name_shirt']

for col in cols:
    data[col] = data[col].astype(str)
    for k, v in mapping.items():
        data[col] = data[col].str.replace(k, v, regex=False)
```

```python
data['name_shirt'] = data['name_shirt'].astype(str)\
.str.encode('latin1', errors='ignore')\
.str.decode('utf-8', errors='ignore')
```
