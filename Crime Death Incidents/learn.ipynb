```python
df = pd.DataFrame({
    'name': ['sara','seif','haytham','arwa'],
    'age': [22,22,37,26],
    'salary': [5000,8000,27000,6000],
    'gender': ['Female','Male','Male','Female']
})
```

```python
df2 = pd.DataFrame({
    'country': ['egypt','egypt','usa','usa'],
    'city': ['cairo','alex','ny','la'],
    'sales': [100,200,300,400]
})
```

```python
fig1 = px.scatter(df, x='age', y='salary')
fig2 = px.bar(df, x='name', y='salary',color='salary')
fig3 = px.line(df, x='age', y='salary')
fig4 = px.histogram(df, y='salary')
fig5 = px.pie(df, names='name', values='salary')
fig6 = px.box(df, y='salary')
fig7 = px.violin(df, y='salary')
fig8 = px.area(df, x='age', y='salary')
fig9 = px.sunburst(df2, path=['country','city'], color='sales')
fig10 = px.treemap(df2, path=['country','city'], values='sales')
fig11 = px.funnel(df, x='salary', y='name')
fig12 = px.density_heatmap(df, x='age', y='salary')
```

```python
fig1.show()  #df1
fig2.show()  #df1
fig3.show()  #df1
fig4.show()  #df1
fig5.show()  #df1
fig6.show()  #df1
fig7.show()  #df1
fig8.show()  #df1
fig9.show()  #df2-----
fig10.show() #df2-----
fig11.show() #df1
fig12.show() #df1
```

```python
fig2 = px.bar(df, x='age', y='salary')
fig5 = px.pie(df, names='name', values='salary')
fig11 = px.funnel(df, x='salary', y='name')

dashboard = make_subplots(
    rows=2,
    cols=2,
    specs=[
        [{"type": "xy"}, {"type": "domain"}],
        [{"type": "xy"}, None]
    ]
)

for trace in fig2.data:
    dashboard.add_trace(trace, row=1, col=1)

for trace in fig5.data:
    dashboard.add_trace(trace, row=1, col=2)

for trace in fig11.data:
    dashboard.add_trace(trace, row=2, col=1)

dashboard.show()
```
