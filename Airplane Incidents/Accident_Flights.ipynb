```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('Airplane_Crashes_and_Fatalities_Since_1908_20190820105639.xlsx')
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
```

```python
data.head(15)
# data.describe()
# data.info()
# data.isnull().sum()
```

```python
data['RAgent'] = data['Operator'].str.split('-').str[0].str.strip()                         #1
data['Operator'] = data['Operator'].str.replace(r'-.*$',"",regex=True)                      
data = data.drop_duplicates()                                                               #4 drop duplicates
data = data.dropna(subset=['Date','Location'])                                              #5 drop nulls in Data , Location
data['Date'] = pd.to_datetime(data['Date'], errors='coerce')                                #6 convert Date to DateTime datatype used for analysis
data['City'] = data['Location'].str.split(',').str[0].str.strip()                           #7 making column city from location by spliting values in column
data['City'] = data['City'].str.title()                                                     #8 make strings in City be title format
data.loc[data['City'].str.contains('Over|Near|Off', case=False, na=False), 'City'] = None   #9 drop NaN contain over | near | off in City 
data['Fatalities'] = data['Fatalities'].fillna(0)                                           #10 fill nulls in Fatalities by 0 
```
